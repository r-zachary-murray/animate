```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.random.normal(0,1,5000)
y = np.random.normal(0,1,5000)
z = np.random.normal(0,1,5000)
c = np.random.uniform(0,1,5000)

angles = np.linspace(0,360,600) #choose 600 viewing angles

def makeplot(i):    
    # Make the plot
    fig = plt.figure(figsize=(4.5,4.5))
    ax = plt.axes(projection='3d')
    ax.scatter(x,y,z, cmap=plt.cm.rainbow, c=c,s=5)
    ax.view_init(30,angles[i])
    plt.axis('off')
    return fig
```
